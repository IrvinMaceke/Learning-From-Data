```python
import pandas as pd

# Load data from CSV file
sa_data = pd.read_csv('Datafiles/stx.csv')
us_data = pd.read_csv('Datafiles/s&p.csv')
jp_data = pd.read_csv('Datafiles/nikkei.csv')
hg_data = pd.read_csv('Datafiles/hangsengindex.csv')
uk_data = pd.read_csv('Datafiles/ftse.csv')

#south africa

# Set the date column as the index
sa_data = sa_data.set_index('timestamp')

# Calculate daily returns
sa_data['daily_return'] = sa_data['close'].astype(float).pct_change()

# Calculate average annual return
sa_annual_return = (sa_data['daily_return'].mean() * 252) * 100

# Calculate annual volatility (standard deviation)
sa_annual_volatility = (sa_data['daily_return'].std() * 252 ** 0.5) * 100

# Print results
print('')
print('')
print('South Africa Satrix Top 40:')
print(f'Average annual return: {sa_annual_return:.2f}%')
print(f'Annual volatility: {sa_annual_volatility:.2f}%')

#United states of america

# Set the date column as the index
us_data = us_data.set_index('timestamp')

# Calculate daily returns
us_data['daily_return'] = us_data[' close'].astype(float).pct_change()

# Calculate average annual return
us_annual_return = (us_data['daily_return'].mean() * 252) * 100

# Calculate annual volatility (standard deviation)
us_annual_volatility = (us_data['daily_return'].std() * 252 ** 0.5) * 100

# Print results
print('')
print('')
print('Unites states Satrix Top 40:')
print(f'Average annual return: {sa_annual_return:.2f}%')
print(f'Annual volatility: {sa_annual_volatility:.2f}%')

#Japan

# Set the date column as the index
jp_data = jp_data.set_index('timestamp')

# Calculate daily returns
jp_data['daily_return'] = jp_data[' close'].astype(float).pct_change()

# Calculate average annual return
jp_annual_return = (jp_data['daily_return'].mean() * 252) * 100

# Calculate annual volatility (standard deviation)
jp_annual_volatility = (jp_data['daily_return'].std() * 252 ** 0.5) * 100

# Print results
print('')
print('')
print('Japan Satrix Top 40:')
print(f'Average annual return: {jp_annual_return:.2f}%')
print(f'Annual volatility: {jp_annual_volatility:.2f}%')

#HongKong

# Set the date column as the index
hg_data = hg_data.set_index('timestamp')

# Calculate daily returns
hg_data['daily_return'] = hg_data[' close'].astype(float).pct_change()

# Calculate average annual return
hg_annual_return = (hg_data['daily_return'].mean() * 252) * 100

# Calculate annual volatility (standard deviation)
hg_annual_volatility = (hg_data['daily_return'].std() * 252 ** 0.5) * 100

# Print results
print('')
print('')
print('Japan Satrix Top 40:')
print(f'Average annual return: {hg_annual_return:.2f}%')
print(f'Annual volatility: {hg_annual_volatility:.2f}%')


#United kingdom

# Set the date column as the index
uk_data = uk_data.set_index('timestamp')

# Calculate daily returns
uk_data['daily_return'] = uk_data[' close'].astype(float).pct_change()

# Calculate average annual return
uk_annual_return = (uk_data['daily_return'].mean() * 252) * 100

# Calculate annual volatility (standard deviation)
uk_annual_volatility = (uk_data['daily_return'].std() * 252 ** 0.5) * 100

# Print results
print('')
print('')
print('United Kingdom Satrix Top 40:')
print(f'Average annual return: {uk_annual_return:.2f}%')
print(f'Annual volatility: {uk_annual_volatility:.2f}%')
print('')
print('')
```

```python
import matplotlib.pyplot as plt
# Calculate daily returns
sa_data['daily_return'] = sa_data['close'].astype(float).pct_change()

# Calculate average annual return
avg_annual_return = (1 + sa_data['daily_return'].mean())**252 - 1

# Calculate volatility
volatility = sa_data['daily_return'].std() * 252 ** 0.5

# Create subplots for the graphs
fig, axs = plt.subplots(2, 2, figsize=(15,10))

# Plot the daily closing prices
axs[0, 0].plot(sa_data['close'])
axs[0, 0].set_xlabel("Date")
axs[0, 0].set_ylabel("Closing Price")
axs[0, 0].set_title("South Africa (SATRIX Top 40) Daily Closing Prices")

# Plot the daily returns
axs[0, 1].plot(sa_data['daily_return'])
axs[0, 1].set_xlabel("Date")
axs[0, 1].set_ylabel("Daily Return")
axs[0, 1].set_title("South Africa (SATRIX Top 40) Daily Returns")

# Plot the annualized returns
axs[1, 0].bar(['Annualized Return'], [avg_annual_return])
axs[1, 0].set_title("South Africa (SATRIX Top 40) Annualized Returns")
axs[1, 0].set_ylabel("Return")

# Plot the volatility
axs[1, 1].bar(['Volatility'], [volatility])
axs[1, 1].set_title("South Africa (SATRIX Top 40) Volatility")
axs[1, 1].set_ylabel("Volatility")

# Display the plots
plt.show()
```
